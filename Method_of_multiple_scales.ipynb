# **Prescribed orientation**

Consider the motion of a self-propelled object in a 2D shear flow. We define the object centre of mass to be $(x(t), y(t))$, and its angular orientation to be $\phi(t)$ (both with respect to laboratory-fixed coordinates). The object translates at (constant) speed $V$ in the direction of its orientation $\phi(t)$. Hence, the governing equations for its centre of mass are:
\begin{align}
\dfrac{\mathrm{d}x}{\mathrm{d}t} = V \cos \phi, \qquad
\dfrac{\mathrm{d}y}{\mathrm{d}t} = V \sin \phi.
\end{align}
We are free to define the origin of the laboratory-fixed coordinate system. For convenience, we define this origin to be the initial position of the object. (We discuss the orientation of these coordinates later when we introduce a flow.) Therefore, the corresponding initial conditions are
\begin{align}
x(0) = 0, \qquad y(0) = 0,
\end{align}
and we note that choosing general initial conditions doesn't change much.

If $\phi(t)$ is prescribed, then the equations above represent a closed system that completely define the motion of the object. However, if $\phi(t)$ varies according to some rules (e.g. prescribed yawing and/or forced by some external flow), then $\phi(t)$ must be determined by some equation of motion in its own right.

# **Jeffery's orbits in 2D**

The evolution of the orientation of a *passive* ellipsoid in shear flow is a famous result in fluid mechanics. These dynamics are known as **Jeffery's orbits**, named after George B Jeffery of UCL who derived them in his classic paper just over 100 years ago [Jeffery, GB. Proc R Soc A (1922), 102(715):161-79]. These theoretical predictions were validated experimentally by Geoffrey I Taylor a year later. [Taylor GI. Proc R Soc A (1923), 103(720):58-61].

While limiting ourselves to shear flow may seem overly restrictive, shear flow is actually the canonical flow to consider if we are interested in single *microswimmer* orientation. This is because the small size of a microswimmer means that it only really cares about the local flow. And locally a general flow will look like a bulk (constant) flow plus a shear flow (consider a Taylor expansion of the flow around any point). Importantly, a bulk flow does not affect orientation since it simply translates the object. Additionally, we can choose the orientation of the laboratory-fixed coordinates so that the $x$-coordinate is in the direction of the flow.

Constraining Jeffery's equations [Jeffery, GB. Proc R Soc A (1922), 102(715):161-79] to 2D motion, the appropriate orientation equation for an object with no self-generated yawing is
\begin{align}
\dfrac{\mathrm{d} \phi}{\mathrm{d}t} = \dfrac{\gamma}{2} \left(1 - B \cos 2 \phi \right).
\end{align}
Here, $\gamma$ is the strength of the externally imposed shear flow, and $B = (r^2 - 1)/(r^2 + 1)$ is a constant that encodes the shape of the ellipsoid through its dependence on $r$, the aspect ratio of the ellipsoid. Specifically, $r>1$ ($0 < B < 1$) corresponds to a prolate spheroid and $0 < r < 1$ ($-1 < B < 0$) corresponds to an oblate spheroid, with $r = 1$ ($B=0$) corresponding to a sphere.

# **Adding self-generated yawing**

We want to specifically consider the effect of a self-generated yawing, so we have to add an activity term to the orientation equation above. A self-generated sinusoidal yawing would have the form $\phi(t) = A \sin \omega t$, where $A$ is the amplitude of yawing. The derivative of this with respect to time (i.e. $A \omega \cos \omega t$) is the rate of change of orientation due to this self-generated yawing. For rapid yawing, we have $\omega \gg 1$.

Given the above, deriving the appropriate orientation equation with self-generated sinusoidal yawing is a straightforward task of adding together the Jeffery and yawing contributions. We can add them together because the microswimmer is acting in a viscous environment where inertia is not important (*Stokes flow*), and so the flow is governed by a system of **linear** PDEs. Viscous effects dominate over inertial effects here (the *Reynolds number* of the flow is small) because microswimmer lengthscales are very small. (See e.g. Lauga, E, & Powers, TR. Reports on Progress in Physics (2009), 72(9): 096601)


Given this, the appropriate orientation equation with self-generated sinusoidal yawing is the sum of the terms discussed above, and hence becomes
\begin{align}
\dfrac{\mathrm{d} \phi}{\mathrm{d}t} = \dfrac{\gamma}{2} \left(1 - B \cos 2 \phi \right) + A \omega \cos \omega t.
\end{align}
The corresponding initial condition for the orientation is
\begin{align}
\phi(0) = \Theta,
\end{align}
where we aren't free to align the coordinates with the initial orientation because we've already orientated them with the direction of the shear flow.

The full system therefore consists of the equation and initial condition immediately above, together with the system discussed above for the position of the object $(x(t),y(t))$.

# **Solving the full dynamics**

We can solve the full system numerically using an in-built ODE solver:

```python
import numpy as np
import math
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parameters
gamma = 1
r = 3
A = 1
w = 20
V = 1
B = (r**2 - 1) / (r**2 + 1)

# Error tolerance
options = {'rtol': 1e-13, 'atol': 1e-13}

# Initial conditions
init = [0, 0, 0]

# Time interval
t_interval = [0, 5]

def full_system_with_flow_and_yawing(t, z, gamma, B, A, w, V):
  return [V*np.cos(z[2]), V*np.sin(z[2]), gamma/2 * (1 - B*np.cos(2*z[2])) + A*w*np.cos(w*t)]

# Solve full system
sol_full = solve_ivp(
    lambda t, z: full_system_with_flow_and_yawing(t, z, gamma, B, A, w, V),
    t_interval,
    init,
    method='LSODA',
    vectorized=True,
    **options
)

# Plot position trajectory
plt.figure()
plt.plot(sol_full.y[0], sol_full.y[1], 'k', linewidth=3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Position Trajectory')
plt.legend(['Full dynamics'])
plt.show()

# Plot orientation dynamics
plt.figure()
plt.plot(sol_full.t, sol_full.y[2], 'k', linewidth=3)
plt.xlabel('Time')
plt.ylabel('Phi')
plt.title('Orientation Dynamics')
plt.legend(['Full dynamics'])
plt.show()
```

# **"Simple" averaging approximation**

Given that the fast yawing rate of $A \omega \cos \omega t$ leads to a fast yawing of $A \sin \omega t$ (which has an average directional bias of zero over one oscillation), we might be tempted to believe that the effect of fast yawing averages out to zero overall effect. That is, we might be tempted to assume that the yawing effect can be ignored and that we would get a good approximation of the emergent dynamics by setting $A=0$ and just simulating the classic Jeffery equation in 2D. Let's check whether this is the case by comparing the full dynamics solved above with the numerical solution when $A = 0$:

```python
# Solve 'simple' averaged system
sol_simple = solve_ivp(
    lambda t, z: full_system_with_flow_and_yawing(t, z, gamma, B, 0, w, V),
    t_interval,
    init,
    method='LSODA',
    vectorized=True,
    **options
)

# Plot position trajectory
plt.figure()
plt.plot(sol_full.y[0], sol_full.y[1], 'k', linewidth=3)
plt.plot(sol_simple.y[0], sol_simple.y[1], 'g:', linewidth=3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Position Trajectory')
plt.legend(['Full dynamics', '"Simple" averaged system'])
plt.show()

# Plot orientation dynamics
plt.figure()
plt.plot(sol_full.t, sol_full.y[2], 'k', linewidth=3)
plt.plot(sol_simple.t, sol_simple.y[2], 'g:', linewidth=3)
plt.xlabel('Time')
plt.ylabel('Phi')
plt.title('Orientation Dynamics')
plt.legend(['Full dynamics', '"Simple" averaged system'])
plt.show()
```

It looks like we need to be a little more careful in how we average.

# **Accurate emergent behaviour (via MoMS)**

From the analysis in lectures, we find that the appropriate emergent behaviour does actually follow the classic Jeffery's equation (i.e. setting $A = 0$ in the self-generated yawing version of Jeffery's equation). However, obtaining the correct behaviour specifically requires appropriate modification of the parameter $B$. The correct effective parameter is $B_{\text{eff}} = J_0(2 A) B$, where $J_0(z)$ is the Bessel function of the first kind, with order zero. Similarly, the correct trajectory is only obtained if the parameter $V$ is modified to $V_{\text{eff}} = J_0(A) V$, using the effective orientation obtained with $B_{\text{eff}}$. Let's simulate and compare what happens when we use these modified parameters:

```python
from scipy.special import jv

# Define effective coefficients
B_effective = B * jv(0, 2 * A)
V_effective = V * jv(0, A)

# Solve reduced system
sol_effective = solve_ivp(
    lambda t, z: full_system_with_flow_and_yawing(t, z, gamma, B_effective, 0, w, V_effective),
    t_interval,
    init,
    method='LSODA',
    vectorized=True,
    **options
)

# Plot position trajectory
plt.figure()
plt.plot(sol_full.y[0], sol_full.y[1], 'k', linewidth=3)
plt.plot(sol_effective.y[0], sol_effective.y[1], 'r--', linewidth=3)
plt.plot(sol_simple.y[0], sol_simple.y[1], 'g:', linewidth=3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Position Trajectory')
plt.legend(['Full dynamics', 'Emergent long-time dynamics', '"Simple" averaged system'])
plt.show()

# Plot orientation dynamics
plt.figure()
plt.plot(sol_full.t, sol_full.y[2], 'k', linewidth=3)
plt.plot(sol_effective.t, sol_effective.y[2], 'r--', linewidth=3)
plt.plot(sol_simple.t, sol_simple.y[2], 'g:', linewidth=3)
plt.xlabel('Time')
plt.ylabel('Phi')
plt.title('Orientation Dynamics')
plt.legend(['Full dynamics', 'Emergent long-time dynamics','"Simple" averaged system'])
plt.show()
```
